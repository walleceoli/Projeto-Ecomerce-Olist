```python
ordens_conjun_dados.isnull().sum()
```

```python
df_nulos = ordens_conjun_dados[
    ordens_conjun_dados['order_approved_at'].isna()]
df_nulos.head()
```

```python
df_nulos_filtro_delive = df_nulos[df_nulos['order_status'] == 'delivered']
df_nulos_filtro_delive
```

### Resultado

**Sobre as colunas do DF**

* Coluna **order_id**:
Representa o identificador unico do pedido.

* Coluna **customer_id**:
Representa o Id de cada pedido.

* Coluna **order_status**:
Representa o status do pedido, se foi enviado, cancelado e etc.

* Coluna **order_purchase_timestamp**:
Representa a data de carimbo da compra  

* Coluna **order_approved_at**:
Representa a data de aprovação do pedido 

* Coluna **order_delivered_carrier_date**:
Representa a data de entrega pela trasportadora

* Coluna **order_delivered_customer_date**:
Representa a data de entrega ao cliete 

* Coluna **order_estimated_delivery_date**:
Representa a data estimada de entrega

**Conclusão**

As colunas de datas estavam com o tipo de dado incorreto, então converti todas para **datetime64[ns]**.

Em seguida, identifiquei que algumas colunas — **order_approved_at, order_delivered_carrier_date e order_delivered_customer_date** — possuíam valores **nulos**. Analisei detalhadamente cada caso para definir o melhor tratamento:

os dados nulos vinham de 3 categorias de status diferentes

* Pedidos **cancelados**:

    Os valores nulos são esperados, pois não faz sentido um pedido cancelado ter datas de aprovação ou entrega.

    **Mantive os nulos**, pois podem ser úteis para **métricas futuras**.

* Pedidos **entregues (delivered)** com order_approved_at nulo:

    Indica **inconsistência**, pois um pedido entregue deveria ter data de aprovação.

    Para resolver, **substituí os nulos** pelo valor de **order_purchase_timestamp**, já que a aprovação geralmente ocorre logo após a compra.

* Pedidos **criados (created)** sem datas:

    Esses pedidos ainda **não foram aprovados ou confirmados**, então é natural que tenham datas nulas.

    **Mantive os nulos**, seguindo a mesma lógica dos pedidos cancelados.

Após essa análise, o DataFrame está com os tipos corretos, os valores nulos tratados de forma contextual e consistente, e pronto para as próximas etapas de análise ou ETL.
